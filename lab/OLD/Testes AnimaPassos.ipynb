```python
# Esse deu certo
%matplotlib notebook

from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.artist import Artist
import numpy as np

def getImage(path, zoom=0.4):
    return OffsetImage(plt.imread(path), zoom=zoom)

x = []
y = []
pD_image_path = '/home/ronie/prog/dnc/lab/pD.PNG'
pE_image_path = '/home/ronie/prog/dnc/lab/pE.PNG'

pD_x0 = 1
pD_y0 = 0


pE_x0 = -1
pE_y0 = 0

figure, ax = plt.subplots()

# Setting limits for x and y axis
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid()

# Inicializa pD e pE
pD_xx = pD_x0
pD_yy = 0

pE_xx = pE_x0
pE_yy = 0



# Insere PÃ© Direito - Vai se Mover
abPD = AnnotationBbox(getImage(pD_image_path), (pD_xx, pD_yy), frameon=False)
ax.add_artist(abPD) 

# Insere PÃ© Esquerdo - Vai ficar parado
abPE = AnnotationBbox(getImage(pE_image_path), (pE_xx, pE_yy), frameon=False)
ax.add_artist(abPE) 

# Since plotting a single graph
line,  = ax.plot(0, 0,'bo') 


def animation_function(i):
    x = i
    y = pD_y0
    
    pD_xx = i  
    pD_yy = pD_y0
    
    line.set_xdata(x)
    line.set_ydata(y)
    abPD.xybox = (pD_xx,pD_yy)
    
    return line,abPD,
  
animation = FuncAnimation(figure,
                          func = animation_function,
                          frames = np.arange(pD_x0 + 1, 6, 0.5), 
                          interval = 100)

animation.save('abre3.mp4', fps=10)


plt.show()
```
