```python
rootCod = db.tbNO_append('root',0,-1,-1,-1,-1,-1,'',True)
with open('ex.csv') as csvfile:
    listPai = []
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    c = 0
    for row in spamreader:
        c = c + 1
        print('Iteração ' + str(c))
        print(row)
        numPai = int(row[0])
        strN = row[1]
        strT = row[2]
        codNo = db.newTAG(strN,strT)

        if numPai == 1:
            #Meu pai é o root
            db.tbAD_append('root-' + strT,rootCod,codNo)
        else:
            db.tbAD_append('tag-tag',listPai[numPai-2],codNo)
        
        
        if len(listPai) < numPai:
            listPai.append(codNo)
        else:
            listPai[numPai-1] = codNo

        print(listPai)
            

db.mostrar()      
```

```python
SQL = '''
SELECT
tbAD.cod as cod,
tbTAG.strNAME as nameFROM,
tbAD.codTo
FROM
tbAD,
tbTAG
Where
tbAD.codFrom = tbTAG.codNo
'''

db.retPandasDfFromSQL(SQL)
```
